```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../googleplaystore.csv')
```

```python
df.shape
```

```python
df
```

```python
def labeling_function(row):
    if pd.isnull(row['Rating']):
        return 'N/A'
    elif row['Rating'] > 4:
        return "High"
    else:
        return "Low"
```

```python
df['Rating_label'] = df.apply(labeling_function, axis=1)
```
