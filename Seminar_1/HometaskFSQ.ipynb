```python
import requests
import json
import pandas
```

```python
# Учётные данные API
client_id = "__"
client_secret = "__"

# Конечная точка API
endpoint = "https://api.foursquare.com/v3/places/search"
```

```python
city = input("Введите город: ")
category= input("Введите интересующую категорию: ")
params = {
    'client_id': client_id,
    'client_secret': client_secret,
    'near': city,
    'query': category
}

headers = {
    'Accept': "application/json",
    'Authorization': "fsq3V3AFHzvqod5PVkb9j5ptfec29VfLTGG2XbHrQEGC8bI="
}
```

```python
response = requests.get(endpoint, params=params, headers=headers)
if response.status_code == 200:
    print(f'Успешный запрос!\n')
    data = json.loads(response.text)
    venues = data['results']
    df = pandas.DataFrame()
    for venue in venues:
        print("Название: ", venue["name"])
        try:
            address = venue["location"]["address"]
        except KeyError:
            address = 'Адрес не указан'
        try:
            rating = venue["rating"]
        except KeyError:
            rating = 'Рейтинг не указан'
        print("Адрес: ", address)
        print("Рейтинг: ", rating)
        print("\n")
        df = pandas.concat([df, pandas.DataFrame([{'Название': venue["name"], 'Адрес': address, "Рейтинг": rating}])], ignore_index = True)
else:
    print(f'Запрос неуспешен, код ошибки - {response.status_code}')
```
